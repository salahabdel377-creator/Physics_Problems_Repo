```python
# Problem 10: Part 4 - Effect of Time Step on Accuracy

# Test different time steps
time_steps = [0.5, 0.2, 0.1, 0.05, 0.02, 0.01]
v_errors_rms = []
a_errors_rms = []

for dt_test in time_steps:
    # Create data with this time step
    t_test = np.arange(0, 10.01, dt_test)
    x_test = x_func(t_test)
    v_analytical_test = v_analytical(t_test)
    a_analytical_test = a_analytical(t_test)
    
    # Numerical velocity
    v_numerical_test = np.zeros_like(t_test)
    v_numerical_test[0] = (x_test[1] - x_test[0]) / dt_test
    v_numerical_test[-1] = (x_test[-1] - x_test[-2]) / dt_test
    for i in range(1, len(t_test)-1):
        v_numerical_test[i] = (x_test[i+1] - x_test[i-1]) / (2*dt_test)
    
    # Numerical acceleration
    a_numerical_test = np.zeros_like(t_test)
    for i in range(1, len(t_test)-1):
        a_numerical_test[i] = (x_test[i+1] - 2*x_test[i] + x_test[i-1]) / (dt_test**2)
    a_numerical_test[0] = (x_test[2] - 2*x_test[1] + x_test[0]) / (dt_test**2)
    a_numerical_test[-1] = (x_test[-1] - 2*x_test[-2] + x_test[-3]) / (dt_test**2)
    
    # Compute RMS errors (excluding boundaries to avoid edge effects)
    v_error_test = v_numerical_test[1:-1] - v_analytical_test[1:-1]
    a_error_test = a_numerical_test[1:-1] - a_analytical_test[1:-1]
    
    v_rms = np.sqrt(np.mean(v_error_test**2))
    a_rms = np.sqrt(np.mean(a_error_test**2))
    
    v_errors_rms.append(v_rms)
    a_errors_rms.append(a_rms)

# Create convergence plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Velocity error vs time step
ax = axes[0]
ax.loglog(time_steps, v_errors_rms, 'b-o', linewidth=2, markersize=8, label='Actual error')
# Reference lines for scaling
dt_ref = np.array(time_steps)
ax.loglog(dt_ref, 0.01*dt_ref**2, 'r--', linewidth=2, alpha=0.7, label='∝ Δt² (2nd order)')
ax.loglog(dt_ref, 0.001*dt_ref, 'g--', linewidth=2, alpha=0.7, label='∝ Δt (1st order)')
ax.set_xlabel('Time Step Δt (s)', fontsize=11)
ax.set_ylabel('RMS Error (m/s)', fontsize=11)
ax.set_title('Velocity Error vs Time Step\n(Convergence Study)', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3, which='both')
ax.legend(fontsize=10)

# Acceleration error vs time step
ax = axes[1]
ax.loglog(time_steps, a_errors_rms, 'r-o', linewidth=2, markersize=8, label='Actual error')
# Reference lines
ax.loglog(dt_ref, 0.01*dt_ref, 'b--', linewidth=2, alpha=0.7, label='∝ Δt (1st order)')
ax.loglog(dt_ref, 0.001, 'g--', linewidth=2, alpha=0.7, label='Constant (0th order)')
ax.set_xlabel('Time Step Δt (s)', fontsize=11)
ax.set_ylabel('RMS Error (m/s²)', fontsize=11)
ax.set_title('Acceleration Error vs Time Step\n(Convergence Study)', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3, which='both')
ax.legend(fontsize=10)

plt.tight_layout()
plt.show()

# Print convergence table
print("\n" + "="*80)
print("CONVERGENCE ANALYSIS: Effect of Time Step on Accuracy")
print("="*80)
print(f"\n{'Δt (s)':>8} | {'N points':>10} | {'V-error RMS (m/s)':>18} | {'a-error RMS (m/s²)':>18}")
print("-"*80)

for i, dt_test in enumerate(time_steps):
    num_points = len(np.arange(0, 10.01, dt_test))
    print(f"{dt_test:8.3f} | {num_points:10d} | {v_errors_rms[i]:18.8f} | {a_errors_rms[i]:18.8f}")

print("\n" + "="*80)
print("CONCLUSIONS:")
print("="*80)
print("""
1. VELOCITY ERROR decreases approximately as Δt² (2nd order accuracy)
   - This is expected for central difference formula
   - Halving Δt reduces error by factor of 4
   
2. ACCELERATION ERROR decreases approximately as Δt (1st order accuracy)
   - Acceleration uses second derivatives, which are less accurate
   - Halving Δt only reduces error by factor of 2
   
3. TRADE-OFFS:
   - Smaller Δt → better accuracy but more computation
   - Too small Δt → numerical noise becomes significant
   - For practical applications: Δt = 0.1 s provides good balance
   
4. OPTIMAL TIME STEP:
   - For this problem: Δt ~ 0.05-0.1 s is recommended
   - Smaller steps don't proportionally improve accuracy (diminishing returns)
   - Larger steps accumulate significant truncation errors
""")
```

## Problem 10: Analysis of Motion from Numerical Data

### Given Data
For the measurement data: $x(t) = t + \frac{1}{20}t^2$ on interval $t \in [0, 10]$ with time step $\Delta t = 0.1$ s

### Part 1: Velocity Approximation (Finite Difference Method)

**Backward difference (2nd order accurate):**
$$v_i \approx \frac{x_{i+1} - x_{i-1}}{2\Delta t}$$

**Forward/Backward at boundaries:**
$$v_0 \approx \frac{x_1 - x_0}{\Delta t}, \quad v_n \approx \frac{x_n - x_{n-1}}{\Delta t}$$

### Part 2: Acceleration Approximation

**Central difference (2nd order accurate):**
$$a_i \approx \frac{x_{i+1} - 2x_i + x_{i-1}}{(\Delta t)^2}$$

### Part 3: Analytical Solution

For $x(t) = t + \frac{1}{20}t^2$:

**Analytical velocity:**
$$v(t) = \frac{dx}{dt} = 1 + \frac{1}{10}t$$

**Analytical acceleration:**
$$a(t) = \frac{dv}{dt} = \frac{1}{10} = 0.1 \text{ m/s}^2 \text{ (constant)}$$

### Part 4: Effect of Time Step on Accuracy

Smaller time steps $\Delta t$ generally improve accuracy but increase computational cost and can amplify numerical noise. We will investigate this trade-off.

## Problem 9: Change of Reference Frame

### Heliocentric System (Sun at the center)

**Earth orbit:**
$$\vec{r}_Z(t) = R_Z(\cos(\omega_Z t), \sin(\omega_Z t))$$

**Mars orbit:**
$$\vec{r}_M(t) = R_M(\cos(\omega_M t), \sin(\omega_M t))$$

where:
- $R_Z \approx 1$ AU (Earth's orbital radius)
- $R_M \approx 1.52$ AU (Mars' orbital radius)
- $\omega_Z = 2\pi$ rad/year (Earth's angular velocity)
- $\omega_M \approx 1.06\pi$ rad/year (Mars' angular velocity)

### Geocentric System (Earth at the center)

The position of Mars as seen from Earth:
$$\vec{r}_{M/Z}(t) = \vec{r}_M(t) - \vec{r}_Z(t) = R_M(\cos(\omega_M t), \sin(\omega_M t)) - R_Z(\cos(\omega_Z t), \sin(\omega_Z t))$$

**Components:**
$$x_{M/Z}(t) = R_M\cos(\omega_M t) - R_Z\cos(\omega_Z t)$$
$$y_{M/Z}(t) = R_M\sin(\omega_M t) - R_Z\sin(\omega_Z t)$$

## Problem 7: 2D Motion with Given Acceleration

### Given Data:
- Acceleration: $\vec{a} = (2, -3)$ m/s²
- Initial velocity: $\vec{v}(0) = (1, 0)$ m/s
- Initial position: $\vec{r}(0) = (0, 0)$ m

### Solution

**Velocity vector** (integrate acceleration):
$$\vec{v}(t) = \vec{v}(0) + \vec{a}t = (1, 0) + (2, -3)t = (1 + 2t, -3t)$$

**Position vector** (integrate velocity):
$$\vec{r}(t) = \vec{r}(0) + \vec{v}(0)t + \frac{1}{2}\vec{a}t^2 = (0, 0) + (1, 0)t + \frac{1}{2}(2, -3)t^2$$

$$\vec{r}(t) = \left(t + t^2, -\frac{3}{2}t^2\right)$$

### Trajectory Equation

From $x = t + t^2$ and $y = -1.5t^2$:

We can eliminate $t$ to find: $y = -1.5(x - t^2) = -1.5x + 1.5t^2$

This gives us a parabolic trajectory.

```python
# Problem 6: Cycloid Motion Visualization
R_cycloid = 1  # radius
omega_cycloid = 1  # angular velocity
num_arches = 2.5  # number of cycloid arches to show

# Time array (two complete rotations = 2 arches)
t_p6 = np.linspace(0, 2*np.pi*num_arches/omega_cycloid, 500)

# Parametric equations for cycloid
x_p6 = R_cycloid * (omega_cycloid*t_p6 - np.sin(omega_cycloid*t_p6))
y_p6 = R_cycloid * (1 - np.cos(omega_cycloid*t_p6))

# Velocity
vx_p6 = R_cycloid * omega_cycloid * (1 - np.cos(omega_cycloid*t_p6))
vy_p6 = R_cycloid * omega_cycloid * np.sin(omega_cycloid*t_p6)
v_mag_p6 = np.sqrt(vx_p6**2 + vy_p6**2)

# Acceleration
ax_p6 = R_cycloid * omega_cycloid**2 * np.sin(omega_cycloid*t_p6)
ay_p6 = R_cycloid * omega_cycloid**2 * np.cos(omega_cycloid*t_p6)
a_mag_p6 = np.sqrt(ax_p6**2 + ay_p6**2)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Cycloid trajectory
ax = axes[0, 0]
ax.plot(x_p6, y_p6, 'b-', linewidth=2.5, label='Cycloid path')

# Mark stopping points (cusps)
for n in range(int(np.ceil(num_arches)) + 1):
    t_cusp = n * 2*np.pi / omega_cycloid
    if t_cusp <= t_p6[-1]:
        x_cusp = R_cycloid * (omega_cycloid*t_cusp - np.sin(omega_cycloid*t_cusp))
        y_cusp = R_cycloid * (1 - np.cos(omega_cycloid*t_cusp))
        ax.plot(x_cusp, y_cusp, 'ro', markersize=8)

# Mark center of rolling circle at several positions
for n in [0, 1, 2]:
    t_center = n * np.pi / omega_cycloid
    if t_center <= t_p6[-1]:
        x_center = R_cycloid * omega_cycloid * t_center
        y_center = R_cycloid
        circle = Circle((x_center, y_center), R_cycloid, fill=False, edgecolor='gray', linestyle='--', alpha=0.3)
        ax.add_patch(circle)

ax.axhline(y=0, color='k', linewidth=0.5)
ax.set_xlabel('x (m)', fontsize=11)
ax.set_ylabel('y (m)', fontsize=11)
ax.set_title(f'Cycloid Trajectory (R={R_cycloid}m)\nRed dots: Cusps (stopping points)', fontsize=12)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10)
ax.set_ylim([-0.3, 2.3])

# Speed vs time
ax = axes[0, 1]
ax.plot(t_p6, v_mag_p6, 'r-', linewidth=2.5, label='$|\\vec{v}(t)|$')
ax.axhline(y=2*R_cycloid*omega_cycloid, color='r', linestyle='--', alpha=0.5, 
           label=f'Max: 2Rω = {2*R_cycloid*omega_cycloid} m/s')

# Mark cusps
for n in range(int(np.ceil(num_arches)) + 1):
    t_cusp = n * 2*np.pi / omega_cycloid
    if t_cusp <= t_p6[-1]:
        ax.plot(t_cusp, 0, 'ro', markersize=6)

ax.set_xlabel('Time t (s)', fontsize=11)
ax.set_ylabel('Speed $|\\vec{v}|$ (m/s)', fontsize=11)
ax.set_title('Speed vs Time (Zero at cusps)', fontsize=12)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10)

# Acceleration magnitude vs time
ax = axes[1, 0]
ax.plot(t_p6, a_mag_p6, 'g-', linewidth=2.5, label='$|\\vec{a}(t)|$')
ax.axhline(y=R_cycloid*omega_cycloid**2, color='g', linestyle='--', alpha=0.5,
           label=f'Constant: Rω² = {R_cycloid*omega_cycloid**2} m/s²')
ax.set_xlabel('Time t (s)', fontsize=11)
ax.set_ylabel('Acceleration $|\\vec{a}|$ (m/s²)', fontsize=11)
ax.set_title('Acceleration Magnitude vs Time (Constant)', fontsize=12)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10)

# Phase space: velocity vector field
ax = axes[1, 1]
ax.plot(x_p6, y_p6, 'b-', linewidth=2.5, label='Cycloid path')

# Plot velocity vectors as phase space
for n, idx in enumerate(np.linspace(0, len(t_p6)-1, 12, dtype=int)):
    x_pos = x_p6[idx]
    y_pos = y_p6[idx]
    vx = vx_p6[idx]
    vy = vy_p6[idx]
    scale = 0.15
    if np.sqrt(vx**2 + vy**2) > 0.01:  # Skip cusps
        ax.arrow(x_pos, y_pos, scale*vx, scale*vy,
                head_width=0.08, head_length=0.05, fc='red', ec='red', alpha=0.6)

ax.axhline(y=0, color='k', linewidth=0.5)
ax.set_xlabel('x (m)', fontsize=11)
ax.set_ylabel('y (m)', fontsize=11)
ax.set_title('Cycloid with Velocity Vectors', fontsize=12)
ax.grid(True, alpha=0.3)
ax.set_ylim([-0.3, 2.3])

plt.tight_layout()
plt.show()

print("Problem 6 - Cycloid Results:")
print(f"Circle radius: R = {R_cycloid} m")
print(f"Angular velocity: ω = {omega_cycloid} rad/s")
print(f"Period (one complete arch): T = 2π/ω = {2*np.pi/omega_cycloid:.2f} s")
print(f"\nSpeed characteristics:")
print(f"  Maximum speed: 2Rω = {2*R_cycloid*omega_cycloid} m/s")
print(f"  Speed at cusps: 0 m/s (the point stops when touching ground)")
print(f"\nAcceleration:")
print(f"  Magnitude (constant): Rω² = {R_cycloid*omega_cycloid**2} m/s²")
```

## Problem 6: Cycloid - Trajectory of a Point on a Rolling Circle

### Given Parametric Equations:
A circle of radius $R$ rolls without slipping along the $x$-axis. A point on the rim traces a cycloid:

$$x(t) = R(\omega t - \sin(\omega t)), \quad y(t) = R(1 - \cos(\omega t))$$

### Part 1: Velocity and Acceleration Vectors

**Velocity:**
$$\vec{v}(t) = R\omega(1 - \cos(\omega t), \sin(\omega t))$$

**Acceleration:**
$$\vec{a}(t) = R\omega^2(\sin(\omega t), \cos(\omega t))$$

### Part 2: Speed and Stopping Points

**Speed:**
$$|\vec{v}(t)| = R\omega\sqrt{2(1 - \cos(\omega t))} = 2R\omega\sin\left(\frac{\omega t}{2}\right)$$

**Stopping moments:** The point stops when $|\vec{v}(t)| = 0$, which occurs when $\sin(\omega t/2) = 0$:
- At times: $\omega t = 0, 2\pi, 4\pi, \ldots$ (i.e., $t = 0, T, 2T, \ldots$ where $T = 2\pi/\omega$)
- These are exactly when the point touches the ground

### Part 3: Maximum Magnitudes

**Maximum velocity:**
$$v_{\text{max}} = 2R\omega \quad \text{(occurs when } \sin(\omega t/2) = 1\text{)}$$

**Maximum acceleration:**

## Problem 4: Circular Motion

### Position, Velocity, and Acceleration Vectors

For circular motion with radius $R$ and angular velocity $\omega$:

**Position vector:**
$$\vec{r}(t) = R(\cos(\omega t), \sin(\omega t))$$

**Velocity vector:**
$$\vec{v}(t) = \frac{d\vec{r}}{dt} = R\omega(-\sin(\omega t), \cos(\omega t))$$

**Acceleration vector:**
$$\vec{a}(t) = \frac{d\vec{v}}{dt} = -R\omega^2(\cos(\omega t), \sin(\omega t))$$

### Magnitudes

$$|\vec{r}(t)| = R \quad \text{(constant)}$$

$$|\vec{v}(t)| = R\omega \quad \text{(constant speed)}$$

$$|\vec{a}(t)| = R\omega^2 \quad \text{(constant magnitude)}$$

### Centripetal Acceleration

Note that $\vec{a}(t) = -\omega^2 \vec{r}(t)$, which shows the acceleration is:
- Always directed toward the center (radially inward)
- Proportional to the position vector
- Has magnitude $a_c = \omega^2 R = \frac{v^2}{R}$

This is the **centripetal acceleration** for circular motion.

### Given Parametric Equations:

From the first equation: $t = \sqrt{\frac{x}{2}}$

**Trajectory equation:** $y = \frac{3\sqrt{2}}{4}x^{3/2}$ or equivalently $y \propto x^{3/2}$

### Part 2: Velocity and Acceleration

**Acceleration magnitude:**
$$|\vec{a}(t)| = \sqrt{16 + 324t^2}$$

### Derivation of Equations of Motion

For a projectile launched with initial velocity $v_0$ at angle $\alpha$ above the horizontal:

**Horizontal component:**
$$x(t) = v_0 \cos(\alpha) \cdot t$$

**Vertical component:**
$$y(t) = v_0 \sin(\alpha) \cdot t - \frac{1}{2}gt^2$$

### Key Results

**Time of flight** (when $y = 0$):
$$0 = v_0 \sin(\alpha) \cdot t - \frac{1}{2}gt^2$$
$$t_{\text{flight}} = \frac{2v_0\sin(\alpha)}{g}$$

**Maximum height** (when $v_y = 0$):
$$h_{\text{max}} = \frac{v_0^2 \sin^2(\alpha)}{2g}$$

**Range** (horizontal distance at time of flight):
$$R = v_0\cos(\alpha) \cdot \frac{2v_0\sin(\alpha)}{g} = \frac{v_0^2 \sin(2\alpha)}{g}$$

**Maximum range occurs at:** $\alpha = 45°$, giving $R_{\text{max}} = \frac{v_0^2}{g}$

## Problem 1: Uniform and Uniformly Accelerated Motion

### Given Equation of Motion:
$$x(t) = x_0 + v_0 t + \frac{1}{2} a t^2$$

### Part 1: Derive velocity and acceleration

The velocity is the first derivative of position:
$$v(t) = \frac{dx}{dt} = v_0 + at$$

The acceleration is the second derivative of position:
$$a(t) = \frac{dv}{dt} = a = \text{constant}$$

### Part 2: Calculations for given parameters
- $x_0 = 0$ m
- $v_0 = 5$ m/s
- $a = -2$ m/s²

**Stopping time** (when $v(t) = 0$):
$$0 = 5 + (-2)t$$
$$t_{\text{stop}} = 2.5 \text{ s}$$

**Maximum displacement** (at stopping time):
$$x_{\text{max}} = 0 + 5(2.5) + \frac{1}{2}(-2)(2.5)^2 = 12.5 - 6.25 = 6.25 \text{ m}$$

**Maximum velocity**: Since $a < 0$, velocity decreases, so maximum velocity occurs at $t = 0$:
$$v_{\text{max}} = 5 \text{ m/s}$$
